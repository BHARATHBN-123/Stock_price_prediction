```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset_train = pd.read_csv('/content/drive/MyDrive/stock market prediction/Quote-Equity-TATAMOTORS-EQ-17-12-2021-to-17-06-2022.csv')
dataset_train.head()
dataset_train.drop(['series '] , axis = 1 , inplace = True)
training_set = dataset_train.iloc[:, 1:2].values
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range=(0,1))
training_set_scaled = sc.fit_transform(training_set)

```

```python
len(training_set_scaled)
```

```python
X_train = []
y_train = []
for i in range(60, len(training_set_scaled)):
  # print(training_set_scaled[i])
  X_train.append(training_set_scaled[i-60:i, 0])
  y_train.append(training_set_scaled[i, 0])
X_train, y_train = np.array(X_train), np.array(y_train)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
y_train
```

```python
X_train
```

```python
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import Dense
```

```python
model = Sequential()
model.add(LSTM(units=50,return_sequences=True,input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1))
model.compile(optimizer='adam',loss='mean_squared_error')
model.fit(X_train,y_train,epochs=100,batch_size=32)
```

```python
#prediction part
dataset_test = pd.read_csv('/content/drive/MyDrive/stock market prediction/Quote-Equity-TATAMOTORS-EQ-17-12-2021-to-17-06-2022.csv')

dataset_test.drop(['series '] , axis = 1 , inplace = True)
real_stock_price = dataset_train.iloc[:, 1:2].values
```

```python
dataset_test.columns
```

```python
dataset_total = pd.concat((dataset_train['OPEN '], dataset_test['OPEN ']), axis = 0)
inputs = dataset_total[len(dataset_total) - len(dataset_test) - 60:].values
inputs = inputs.reshape(-1,1)
inputs = sc.transform(inputs)
X_test = []
for i in range(60, len(dataset_total)):
  X_test.append(inputs[i-60:i, 0])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
predicted_stock_price = model.predict(X_test)
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```
